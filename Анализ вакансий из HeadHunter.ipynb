3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)

* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
# hh_df['Опыт работы']
def get_exp(feature):
    if (feature is np.nan) or (feature == 'Не указано'):
        return np.nan
    month_key_words = ['месяц', 'месяцев', 'месяца']
    year_key_words = ['год', 'лет', 'года']
    splitted = feature.split()[0:6]
    month = 0
    year = 0
    for i in range(len(splitted)):
        if splitted[i] in month_key_words:
            month = splitted[i-1]
        if splitted[i] in year_key_words:
            year = splitted[i-1]         
    return int(year)*12 + int(month)

hh_df['Опыт работы (месяц)'] = hh_df['Опыт работы'].apply(get_exp)

hh_df['Опыт работы (месяц)'].median() #для решения заданий модуля

hh_df.drop('Опыт работы', axis=1, inplace=True)
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 'Омск', 'Самара', 
                  'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград']          

def city_set(feature):
    splitted = feature.split(sep=' ,')
    if splitted[0] == 'Москва':
        return splitted[0]
    if splitted[0] == 'Санкт-Петербург':
        return splitted[0]
    if splitted[0] in million_cities:
        return 'город-миллионник'
    else:
        return 'другие'
    
hh_df['Город'] = hh_df['Город, переезд, командировки'].apply(city_set)

hh_df['Город'].value_counts() #для решения заданий модуля
hh_df['Город'].value_counts(normalize=True)

def remove_set(feature):
    splitted = feature.split(sep=',')
    if splitted[1].split()[0] == 'м.': #проверка на наличие метро
        if splitted[2].split()[0] == 'не':
            return False
        else:
            return True
    if splitted[1].split()[0] == 'не':
        return False
    else:
        return True

hh_df['Готовность к переезду'] = hh_df['Город, переезд, командировки'].apply(remove_set)

hh_df['Готовность к переезду'].value_counts() #для задания модуля
    
def trip_set(feature):
    if 'командировка' in feature:
        if ('не готов к командировкам' in feature) or ('не готова к командировкам' in feature):
            return False
        else: 
            return True
    else:
        return False

hh_df['Готовность к командировкам'] = hh_df['Город, переезд, командировки'].apply(trip_set)
    
hh_df.drop('Город, переезд, командировки', axis=1, inplace=True) #удаление дублирующего признака

hh_df['Готовность к командировкам'].value_counts() #для задания модуля
mask = (hh_df['Готовность к командировкам'] == True) & (hh_df['Готовность к переезду'] == True)
hh_df[mask].shape[0] / hh_df.shape[0]
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
# hh_df['Занятость']
 
def full_busy_set(feature):
    splitted = feature.split(sep=',')
    for elem in splitted:
        if elem.split()[0] == 'полная':
            return True
    return False     

def part_busy_set(feature):
    splitted = feature.split(sep=',')
    for elem in splitted:
        if elem.split()[0] == 'частичная':
            return True
    return False

def project_busy_set(feature):
    splitted = feature.split(sep=',')
    for elem in splitted:
        if elem.split()[0] == 'проектная':
            return True
    return False

def intern_busy_set(feature):
    splitted = feature.split(sep=',')
    for elem in splitted:
        if elem.split()[0] == 'стажировка':
            return True
    return False

def volunteer_busy_set(feature):
    splitted = feature.split(sep=',')
    for elem in splitted:
        if elem.split()[0] == 'волонтерство':
            return True
    return False

hh_df['Полная занятость'] = hh_df['Занятость'].apply(full_busy_set)
hh_df['Частичная занятость'] = hh_df['Занятость'].apply(part_busy_set)
hh_df['Проектная работа'] = hh_df['Занятость'].apply(project_busy_set)
hh_df['Cтажировка'] = hh_df['Занятость'].apply(intern_busy_set)
hh_df['Волонтёрство'] = hh_df['Занятость'].apply(volunteer_busy_set)

def full_graph_set(feature):
    splitted = feature.split(sep=',')
    for elem in splitted:
        if elem.split()[1] == 'день':
            return True
    return False

def change_graph_set(feature):
    splitted = feature.split(sep=',')
    for elem in splitted:
        if elem.split()[0] == 'сменный':
            return True
    return False

def flex_graph_set(feature):
    splitted = feature.split(sep=',')
    for elem in splitted:
        if elem.split()[0] == 'гибкий':
            return True
    return False

def distant_graph_set(feature):
    splitted = feature.split(sep=',')
    for elem in splitted:
        if elem.split()[1] == 'работа':
            return True
    return False

def watcher_graph_set(feature):
    splitted = feature.split(sep=',')
    for elem in splitted:
        if elem.split()[1] == 'метод':
            return True
    return False

hh_df['Полный день'] = hh_df['График'].apply(full_graph_set)
hh_df['Сменный график'] = hh_df['График'].apply(change_graph_set)
hh_df['Гибкий график'] = hh_df['График'].apply(flex_graph_set)
hh_df['Удалённая работа'] = hh_df['График'].apply(distant_graph_set)
hh_df['Вахтовый метод'] = hh_df['График'].apply(watcher_graph_set)

task_mask = (hh_df['Проектная работа'] == True) & (hh_df['Волонтёрство'] == True) #для решений заданий модуля
hh_df[task_mask].shape[0]
quest_mask = (hh_df['Вахтовый метод'] == True) & (hh_df['Гибкий график'] == True)
hh_df[quest_mask].shape[0]

hh_df.drop(['Занятость', 'График'], axis=1, inplace=True)
```

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
forex_data = pd.read_csv('data/ExchangeRates.csv')
forex_df = forex_data.copy()
#приводим к формату datetime, а затем выделяем только дату, на выходе - object
hh_df['Обновление резюме'] = pd.to_datetime(hh_df['Обновление резюме'])
hh_df['Обновление резюме'] = hh_df['Обновление резюме'].dt.date
forex_df['date'] = pd.to_datetime(forex_df['date'])

#приводим к типу object, чтобы получилось объединить
forex_df['date'] = forex_df['date'].dt.date
#создаем признак Валюта для последующего объединения по данному столбцу
hh_df['Валюта'] = hh_df['ЗП'].apply(lambda x: x.split()[1])

#переводим валюты в стандарт ISO, при этом ряд валют в переводе не нуждаются
def exchange(currency):
    not_to_swap = ['руб.', 'USD', 'KZT', 'EUR', 'KGS', 'AZN']
    if currency in not_to_swap:
        return currency
    elif currency == 'бел.руб.':
        return 'BYN'
    elif currency == 'грн.':
        return 'UAH'
    else:
        return 'UZS'
    
hh_df['Валюта'] = hh_df['Валюта'].apply(exchange)
hh_df['Валюта'].value_counts()

#объединяем таблицы по двум ключевым столбцам
hh_df = hh_df.merge(forex_df, how='left', left_on=['Обновление резюме', 'Валюта'], right_on=['date', 'currency'])
#заменяем пропуски, так как инфомация о Рублях ожидаемо отсутствовала в таблицe по валютам
values = {
    'close': 1,
    'proportion': 1
}
hh_df = hh_df.fillna(value=values)
#оставляем только числовое значение ожидаемой зарплаты
hh_df['ЗП'] = hh_df['ЗП'].apply(lambda x: x.split()[0])
hh_df['ЗП'] = hh_df['ЗП'].astype('int64')
#cоздаём новый признак
hh_df['ЗП(руб)'] = hh_df['ЗП'] * hh_df['close'] / hh_df['proportion']

hh_df['ЗП(руб)'].median() #для решения задания

hh_df.drop(['ЗП', 'Валюта', 'currency', 'per', 'date', 'time', 'close', 'vol', 'proportion'], axis=1, inplace=True)
hh_df.columns
```

КОММЕНТАРИИ: Справа от моды наблюдается больше наблюдений, нежели слева. Таким образом, распределение имеет правостороннюю асимметрию.

```python
def outliers_z_score(data, left=3, right=4):
    mu=data.mean()
    sigma=data.std()
    lower_bound = mu-left*sigma
    upper_bound = mu+right*sigma
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    cleaned = data[(data > lower_bound) & (data < upper_bound)]
    return outliers, cleaned    

outliers, cleaned = outliers_z_score(log_data)
outliers.values
log_mask = (np.log(hh_df['Возраст']) == outliers.values[0]) | (np.log(hh_df['Возраст']) == outliers.values[1]) | (np.log(hh_df['Возраст']) == outliers.values[2])
hh_df[log_mask]
```
